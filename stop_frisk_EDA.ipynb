```python
# Establishing and cleaning target variable
# In this case, I'm going to create a binary target variable that indicates whether an individual was arrested or
# summoned (1) vs. neither (0). I'm combining arrest and summons summons often result in arrests.

# It should also be noted that there are 18 instances where an individual was arrested AND issued a summons to 
# appear in court. Each of these instances are counted as 1 in our target column.
conditions = [
    df.arstmade==1,
    df.sumissue==1
]

choices = [
    1,
    1
]

df['target'] = np.select(conditions, choices, default=0)
```

```python
def make_stacked_comparison(x1, x2, title, dist, width, rot, position):
    fig, ax = plt.subplots(figsize=(width,6))
    fig.patch.set_facecolor('white')

    y = pd.crosstab(df[x1], df[x2], normalize='index').sort_values(by=df[x1].unique()[0], axis=1, ascending=True)
    x = []
    for each in y.index:
        x.append('%s (n=%s)'%(each, str(df[x1].value_counts().loc[each])))

    colors =["#155F83", "#ea8171", "#8F3931", "#8A9045", "#D5802B", "#350E20", "#58593F", "#880d1e"]

    handles = []
    bottom = 0
    for i in range(y.shape[1]):
        handles.append(ax.bar(x, y.iloc[:,i], bottom = bottom, label = y.index, color=colors[i], width=.6))
        bottom += y.iloc[:,i]

    spine_position = ['top', 'left', 'right']
    for spine in spine_position:    
        ax.spines[spine].set_color('white')

    ax.set_title(title, fontsize=14)
    ax.set_xticklabels(x, rotation=rot, ha=position)
    labels=y.columns

    ax.legend(handles[::-1], labels[::-1], bbox_to_anchor=(dist, .77, 0.3, 0.2), loc='upper right', 
              frameon=False)
    plt.show()
```

### Working first with dummy features

```python
stats_df = pd.get_dummies(stats_df, columns=['pct', 'inout', 'trhsloc', 'typeofid', 'sex', 'race', 'build', 'city', 'month',
                                  'descr'], drop_first=True)
```

```python
# Target column here is a binary variable indicating whether an individual was either issued a summons or arrested 
# (1) or not arrested or issued a summons (0).

# Conducting 2 sample proportion comparisons for dummy variables
dummy_df = stats_df.drop(['perobs', 'perstop', 'repcmd', 'revcmd', 'age', 'weight', 'ht_inches'], axis=1)

dummy_prop = dummy_df.groupby(['target']).sum().T
```

```python
dummy_prop.astype(float)
```

```python
# Difference of 2 proportions Z test for each feature
# H0: p0 = p1
# H1: p0 != p1
# The above hypothesis applies to each feature in the df
from statsmodels.stats.proportion import proportions_ztest

results = []
for each in dummy_prop.index:
    a = dummy_prop[0][each]
    b = dummy_prop[1][each]
    if a!=0 and b!=0:
        count = [a, b]
        nobs = [dummy_df.target.value_counts()[0], dummy_df.target.value_counts()[1]]
        stat, pval = proportions_ztest(count, nobs)
        results.append({'test_stat': round(stat,4), 'p-value': round(pval,4)})
    else:
        results.append({'test_stat': np.nan, 'p-value': np.nan})
```

```python
stat_results = pd.DataFrame(results)
```

```python
stat_results.set_index(dummy_prop.index, inplace=True)
```

```python
# Reconfigure df_prop to show perentages
dummy_prop = dummy_df.groupby(['target']).mean().T
```

```python
dummy_significance = pd.merge(dummy_prop, stat_results, left_index=True, right_index=True)
```

```python
dummy_significance.sort_values(by='p-value', ascending=False)
```

```python
# Run through the above list and for each feature with a p-value above 0.05, add it to a list of non-significant
# features. These will later be removed.

dummies_to_remove = dummy_significance.sort_values(by='p-value', ascending=False).loc[:'month_September',:].append(
    dummy_significance.sort_values(by='p-value', ascending=False).loc['riflshot':'descr_UNLAWFUL ASSEMBLY',:])

# For each feature in this list we are unable to reject the null hypothesis that the proportion of ones
# is equal in both the arrested and not arrested groups.
```

```python
# 129 features were insignificant. NaNs were removed as well because there were very few positive cases in either
# group and most did not have a single case in the other group. As such, we were unable to calculte the difference
# in proportions for these features.

dummies_to_remove.shape
```

```python
# let's compile the list of relevant features for modeling
dummies_to_keep = dummy_significance.sort_values(by='p-value', ascending=False).loc[
    'pct_20':'cs_descr',:]
dummies_to_keep.shape
# There are 97 dummy features that were statistically significant
```

```python
cont_df = stats_df.loc[:,['age', 'weight', 'ht_inches', 'target']]
```

```python
cont_0 = cont_df[cont_df.target==0].drop(['target'], axis=1)
```

```python
cont_1 = cont_df[cont_df.target==1].drop(['target'], axis=1)
```

```python
# 2 sample t-test
# H0: m0 = m1
# H1: m0!= m1

cont_results = []

for each in cont_0.columns:
    a = cont_0[each]
    b = cont_1[each]
    t, p = stats.ttest_ind(a, b, equal_var=False)
    cont_results.append({'feature':each, 'test_stat':t, 'p-value':p})
cont_results_df = pd.DataFrame(cont_results)
```

```python
cont_results_df.set_index('feature', inplace=True)
```

```python
cont_df_means = cont_df.groupby(['target']).mean().T
```

```python
cont_performance = pd.merge(cont_df_means, cont_results_df, left_index=True, right_index=True)
cont_performance
```

```python
cont_features_to_keep = cont_performance.loc['age':'weight', :]
```

```python
cont_features_to_drop = cont_performance.loc['ht_inches',:]
```

```python
# Visualizing number of cases resulting in police force

fig, ax = plt.subplots(figsize=(15,6))
fig.patch.set_facecolor('white')

title = 'Incidence of Police Use-of-Force - Whole Sample'

y = round(stats_df.loc[:,['pf_hands', 'pf_wall', 'pf_grnd', 'pf_drwep', 'pf_ptwep', 'pf_baton', 'pf_hcuff',
                          'pf_pepsp', 'pf_other']].mean(), 3)

feature_names = ['Use of hands', 'Suspect Against Wall', 'Suspect on Ground', 'Weapon Drawn', 'Weapon Pointed',
                 'Baton Used', 'Handcuffs Used', 'Pepperspray Used', 'Other Use of Force']
x = []
for i in range(len(feature_names)):
    x.append('%s (n=%s)'%(feature_names[i], str(pf_counts[i])))

colors =["#D5802B", "#155F83", "#8F3931", "#8A9045", "#ea8171", "#350E20", "#58593F", "#880d1e", "#a63446"]

ax.bar(x, y, color=colors)

spine_position = ['top', 'left', 'right']
for spine in spine_position:    
    ax.spines[spine].set_color('white')
    
for a,b in zip(x,y):

    label = "{:.2f}".format(b)

    plt.annotate(label, # this is the text
                 (a,b), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

ax.set_xticklabels(x, rotation=45, ha='right')
ax.set_title(title, fontsize=14)

plt.show()
```

```python
# Arrest Outcome by whether the person was searched

x1='searched'
x2='arrest_viz'
title='Arrest Outcome by Searched (0/1)'
dist=.8
width=15
position='center'
fig, ax = plt.subplots(figsize=(width,6))
fig.patch.set_facecolor('white')

y = pd.crosstab(df[x1], df[x2], normalize='index')
x = []
for each in y.index:
    x.append('%s (n=%s)'%(each, str(df[x1].value_counts().loc[each])))

colors =["#155F83", "#ea8171", "#8F3931", "#8A9045", "#D5802B", "#350E20", "#58593F", "#880d1e"]

handles = []
bottom = 0
for i in range(y.shape[1]):
    handles.append(ax.bar(x, y.iloc[:,i], bottom = bottom, label = y.index, color=colors[i], width=.6))
    bottom += y.iloc[:,i]

spine_position = ['top', 'left', 'right']
for spine in spine_position:    
    ax.spines[spine].set_color('white')

ax.set_title(title, fontsize=14)
ax.set_xticklabels(x, ha=position)
labels=y.columns

ax.legend(handles[::-1], labels[::-1], bbox_to_anchor=(dist, .77, 0.3, 0.2), loc='upper right', 
          frameon=False)
plt.show()
```
